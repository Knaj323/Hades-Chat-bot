```python
import requests

def get_wikipedia_content(titles):
    url = "https://en.wikipedia.org/w/api.php"
    params = {
        "action": "query",
        "format": "json",
        "prop": "extracts",
        "titles": "|".join(titles)}
    response = requests.get(url, params=params)
    data = response.json()
    
    content_dict = {}
    
    
    for title in titles:
        page_id = next(iter(data['query']['pages']))
        content = data['query']['pages'][page_id]['extract']
        content_dict[title] = content
    
    return content_dict

titles = [ "Hades", "Hades_in_popular_culture"]

content_dict = get_wikipedia_content(titles)

for title, content in content_dict.items():
    print(f"Content for '{title}':")
    print(content)
    print("\n---\n")  # Separator between pages
```
